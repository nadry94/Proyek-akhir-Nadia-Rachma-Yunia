# Heart Disease

1. Columns
2. age age in years
3. sex = (1 = male; 0 = female)
4. cp = chest pain type
5. trestbps = resting blood pressure (in mm Hg on admission to the hospital)
6. chol = serum cholestoral in mg/dl
7. fbs = (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
8. restecg = resting electrocardiographic results
9. thalach = maximum heart rate achieved
10. exang = exercise induced angina (1 = yes; 0 = no)
11. oldpeakST= depression induced by exercise relative to rest
12. slope = the slope of the peak exercise ST segment
13. ca = number of major vessels (0-3) colored by flourosopy
14. thal = 1 = normal; 2 = fixed defect; 3 = reversable defect
15. target = 1 or 0
>0 = no disease and 1 = disease.

>chest pain type expalanation

1 = typical angina
2 = atypical angina
3 = non — anginal pain
4 = asymptotic

## kodingan

```python
from sklearn.model_selection import train_test_split
```

```python
X = jtg.drop('target',axis=1)
y = jtg['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state = 101)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Prediction & Evaluation

```python
predictions = dtree.predict(X_test)
```

```python
predProb = dtree.predict_proba(X_test)
```

```python
dtree.classes_
```

```python
predProb
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
con = pd.DataFrame(data=confusion_matrix(y_test, predictions), columns= ['P sehat', 'P sakit'],index= ['A sehat', 'A sakit'])
print(con)
```

```python
y_test.value_counts()
```

##### sehat = NO, sakit = Yes

>1. A sehat - P sehat = TN (TRUE NEGATIVE)
2. A sakit - P sakit = TP (TRUE POSITIVE)

>3. A sehat -  P sakit = FP (FALSE POSITIVE)
4. A sakit - P sehat = FN (FALSE NEGATIVE)

##### 1. Precision 
1. kemampuan classifier untuk tidak melabeli data yg sebenarnya negatif menjadi positif
2. Precision  Accuracy of positive predictions.
> 1. Precision (target 1 ) = $TP/(TP + FP)$
  2. Precision (target 0) = $TN /(TN+FN)$

##### 2. Recall
1. kemampuan classifier untuk mecari semua yang positif. 
2. Recall: Fraction of positives that were correctly identified.
 >For each class it is defined as the ratio of true positives to the sum of true positives and false negatives.
  1. Recall (target 1) = $TP/(TP+FN)$
  2. Recall (target 0) = $TN/(TN+FP)$

##### 3. F1 Score
1. What percent of positive predictions were correct
2. that the best score is 1.0 and the worst is 0.0. 
> 1. F1 Score (target 1)= $2*(Recall * Precision) / (Recall + Precision)$
  2. F1 Score (target 0)= $2*(Recall * Precision) / (Recall + Precision)$
  
##### 4. Macro average 
1. jumlah nilai precission dibagi jumlah kelas target
